```python
import pandas as pd
import numpy as np

data = {
    'product': ['laptop', 'mobile', 'tablet', 'headphones', 'smartwatch','camera'],
    'price': [55000,20000,30000,2000,8000,45000],
    'unit': ['pcs', 'pcs', 'pcs', 'pcs', 'pcs', 'pcs']
}
df = pd.DataFrame(data)
print(df)
```

```python
# data binning
bins1 = [0,10000,30000,60000]
labels1 = ['low','medium','high']

df['price_category'] = pd.cut(df['price'],bins=bins1,labels=labels1)
print(df)
```

```python
# binning units sold
bins2 = [0,30,60,100]
labels2 = ['low sales','medium sales','high sales']

# Add a 'units_sold' column to the DataFrame to resolve the KeyError
df['units_sold'] = [15, 40, 25, 5, 10, 50] # Example data for units sold

df['sales_category'] = pd.cut(df['units_sold'], bins=bins2 , labels=labels2)
print(df)
```

```python
# data formatting
# convert data type
print(df.dtypes)
df['units_sold'] = df['units_sold'].astype(float)
print(df.dtypes)
df
```

```python
# covert text to uppercase
df['product'].str.upper()
df
```

```python
# rounding values
df['price']= df['price'].round(2)
print(df)
```

```python
# additional operations
# sort data
sorted_df = df.sort_values(by='price', inplace=True, ascending=False)
df
```

```python
# display unique values
print(df['price_category'].unique())
```

```python
# second dataset
import pandas as pd
import numpy as np
data = {
    'order_id':['01','02','03','04','05','06','07','08'],
    'order_value':[250,450,120,600,300,150,500,220],
    'delivery_time':[30,45,25,60,35,20,50,28],
    'distance_km':[5,8,3,10,6,2,9,4]
}
df = pd.DataFrame(data)
print(df)
```

```python
#binning order value
bins = [0,200,400,700]
labels = ['low order','medium order','high order']

df['order_category'] = pd.cut(df['order_value'],bins=bins,labels=labels)
print(df)
```

```python
# binning delivery time
bins2 = [0,30,45,70]
labels2 = ['fast','normal','slow']

df['delivery_speed'] = pd.cut(df['delivery_time'],bins=bins2,labels=labels2)
print(df)
```

```python
bins3 = [0,4,8,10]
labels3 = ['short distance','medium distance','high distance']
df['distance_category'] = pd.cut(df['distance_km'],bins=bins3,labels=labels3)
print(df)
```

```python
# COUNT OF CATEGORIES
print(df['order_category'].value_counts())
df
```

```python
# convert order_id to uppercase
df['order_category'] = df['order_category'].str.upper()
df
```

```python
# sort values
sorted_df = df.sort_values(by='order_value', inplace=True, ascending=False)
df
```
